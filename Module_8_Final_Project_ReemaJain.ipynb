```python
# importing necessary libraries & adding new column 'year' & 'month'
import pandas as pd
import matplotlib.pyplot as plt

df_data = pd.read_csv("mrtssales92-present.csv")
df_data['year'] = pd.DatetimeIndex(df_data['Period']).year
df_data['month'] = pd.DatetimeIndex(df_data['Period']).month
df_data
```

#### 1. Retail and food services sales

```python
# Selecting retail and food services data & used groupby on year to get combined value for years

df_retail=df_data.loc[df_data['Kind_of_Business']=='Retail and food services sales, total']
retail=df_retail.groupby(df_retail['year'],as_index=False).sum()
retail
```

```python
# Plotting Retail and food services sales from 1992-2020


plt.figure(figsize=(10,8))
plt.grid(True)
plt.plot(retail['year'], retail['value'],linestyle='-', linewidth='2')
plt.xlabel("Year")
plt.ylabel("Amount")
plt.title("Retail and food services sales, total")
```

Above graph displays clear trend for 'Total sales for retail and food services'. As per above chart, 'Total sales for retail and food services' followed an increasing trend from 1992-2020 with slight dip in 2009. 

This shows there has been increasing demand for this business. The reasons can be increasing population, increase in per capita income of the people etc. 

Dip in 2009 can be due to economic recession.


#### 2. Book stores,Sporting goods stores ,and Hobby, toy &  game stores 

```python
# Selecting Book stores,Sporting goods stores ,and Hobby, toy &  game stores data. Used groupby on year to get combined value for years

df_book=df_data.loc[df_data['Kind_of_Business']=='Book stores']
book=df_book.groupby(df_book['year'],as_index=False).sum()

df_sport=df_data.loc[df_data['Kind_of_Business']=='Sporting goods stores']
sport=df_sport.groupby(df_sport['year'],as_index=False).sum()

df_toy=df_data.loc[df_data['Kind_of_Business']=='Hobby, toy, and game stores']
toy=df_toy.groupby(df_toy['year'],as_index=False).sum()




```

```python
# Plotting Book stores,Sporting goods stores ,and Hobby, toy &  game stores from 1992-2020

plt.figure(figsize=(10,8))
plt.grid(True)
plt.plot(book['year'], book['value'],linestyle='-', linewidth='2',color='orange',label='book')
plt.plot(sport['year'], sport['value'],linestyle='--', linewidth='2',color='r',label='sport')
plt.plot(toy['year'], toy['value'],linestyle='-.', linewidth='2',color='g',label='hobby_toy_game')
plt.xlabel("Year")
plt.ylabel("Amount")
plt.title("Spendings")
plt.legend()

```

Above graph depicts spendings for 'Book stores,Sporting goods stores ,and Hobby, toy &  game stores' from 1992 to 2020. 

Spending on sports displaying an upward trend as it has been increasing since 1992. This may be due to below reasons:
- Increased health awareness amongst people
- Increased spending capacity leading to increased spending on sports equipments

Spending on books increased from 1992-2007 & then started declining. This decline can be attributred to below:
- Increase in e-book
- According to world economic forum, reading for pleasure has declined in US 
- Increase in other source of entertainment such as online TV(Netflix, Hulu etc.), youtube, social media etc.

Spending on hobbies, toys & games stores shows that spending increased at very slow rate from 1992-2000 with marginal decline from 2000-2010 & then started increasing again with slight decrease in 2019.  
- This may be due to advent of online games.

### 3.4 Explore Percentage Change

Describe which *queries* you wrote to explore the differences in trends between various categories in your data.

In your submission make sure to answer the following:

- In economics, what is the percentage change and why is it considered an important measure to predict quantities like spending patterns?
- Consider the women's clothing and men's clothing businesses and their percentage change. How are these two businesses related? For each of the two businesses, what is the percentage of contribution to the whole and how does it change over time?

Percentage change tells the percent change from previous value over a period of time.Percent change charts are useful when comparing price movements.

```python
# PERCENT CHANGE FROM 1992-2020 for Women's & Men's clothing
# Selecting Women's & Men's clothing data, used groupby on year to get combined value for years & then calculating percent change on values

cloth_woman=df_data[df_data['Kind_of_Business']=='Women\'s clothing stores']
woman=cloth_woman.groupby(cloth_woman['year'],as_index=False).sum()
woman_perc=woman['value'].pct_change()*100
#woman_perc

cloth_man=df_data[df_data['Kind_of_Business']=='Men\'s clothing stores']
man=cloth_man.groupby(cloth_man['year'],as_index=False).sum()
man_perc=man['value'].pct_change()*100
#man_perc


```

```python
# Plotting Women's & Men's clothing from 1992-2020

plt.figure(figsize=(10,8))
plt.grid(True)
plt.plot(woman['year'], woman_perc,linestyle='-', linewidth='2',color='g',label='Women Clothing')
plt.plot(man['year'], man_perc,linestyle='--', linewidth='2',color='r',label='Men Clothing')
plt.xlabel("Year")
plt.ylabel("Percent Change")
plt.title("Clothings")
plt.legend(loc=3)
```

Above graph shows percentage change for women's & men's clothing from 1992-2020. It appears that men & women's clothings dropped tremendously during 2020. One of the reason may be due to pandemic.Excluding 2020, it looks like men & women trend looks pretty similar except in year between 1997-1999. In this period, men's clothing dropped(~-5%) while women's clothing continued increasing at around 6%.

```python
# Calculated  percentage of contribution to men's & women's clothing

cloth_total=df_data[(df_data['Kind_of_Business']=='Women\'s clothing stores') | (df_data['Kind_of_Business']=='Men\'s clothing stores')]
#cloth_total
total=cloth_total.groupby(by=['year', 'Kind_of_Business']).sum()
#total
perc_contr=(total['value']/total.groupby('year')['value'].transform('sum'))*100
perc_contr.unstack(level=1)
```

```python
# Plotting Percentage Contribution for Women's & Men's clothing from 1992-2020

perc_contr.unstack(level=1).plot(kind='line', subplots=False)
```

Above graph clearly depicts that percentage contribution of women's clothing is way higher than that of men's clothing.
Percentage contribution of Men's clothing is declining from approx. 1998.
This shows that women's clothing is a bigger market in clothing industry.

Rolling time window helps in smoothen the highly fluctuating data. You can use shorter rolling window size or larger size depending on data. For data collected in short intervals, short rolling window size will be prefered. For data collected in longer intervals,larger rolling window size is useful.Larger rolling window sizes tends to smoothen trend  more than rolling window with shorter sizes.

In case of MRTS, spending patterns shows monthly fluctuations & seasonality. Rolling time windows can be useful in predicting long term trends of spending patterns as it smoothen out the fluctuations.



Lets, look at rolling averages for Two Business: Paint and wallpaper stores & Jewelry stores. I analyzed data for 3 years(2015-2020).

#### 1. Jewelry stores

```python
#Rolling window for Jewelry stores
# Selecting Jewelry stores data, sortvalues by Period & selected year >=2015

df_jewelry=df_data[df_data['Kind_of_Business']=='Jewelry stores'].copy()
df_jewelry.sort_values(by='Period', inplace=True)
jewelry=df_jewelry[df_jewelry['year']>=2015].copy()
jewelry
```

```python
#calculating Rolling average for 12 & 6 sizes

jewelry['RA_12']=jewelry['value'].rolling(12).mean()
jewelry['RA_6']=jewelry['value'].rolling(6).mean()
jewelry
```

```python
# Plotting monthly sales for Jewelry stores from 2015-2020
fig=plt.figure(figsize=(30,20))
plt.grid(True)
plt.plot(jewelry['Period'], jewelry['value'],linestyle='-', linewidth='2',label='Jewelry Stores')
plt.plot(jewelry['Period'], jewelry['RA_12'],linestyle='--', linewidth='2',label='RA 12 months')
plt.plot(jewelry['Period'], jewelry['RA_6'],linestyle='-.', linewidth='2',label='RA 6 months')
plt.xlabel("Period")
plt.ylabel("Amount")
plt.legend(loc=2)
fig.autofmt_xdate()
```

Above graph shows seasonal variations for jewelry sales. Sales increase during holiday seasons i.e. sales starts increasing in Nov(Thankgiving) of every year & is at peak in December(Christmas,New Year etc.). 

RA_6 (Rolling Average for 6 months) & RA_12(Rolling Average for 12 months) smoothen out the fluctuations in jewelry sales. Looking at rolling averages for 12 months, it looks like sales of jewelry is pretty stable.
